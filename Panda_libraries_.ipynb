```python
# prompt: can you create for all the columns in data

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

data = pd.read_csv('/content/Universities.csv')
data.head()
data.info()

# Loop through all columns in the DataFrame
for column in data.columns:
    # Create a histogram for each column
    sns.histplot(data[column])
    plt.title(f'Distribution of {column}')
    plt.show()

    # Attempt to create a scatter plot if there's a numerical column to compare with
    for other_column in data.columns:
        if column != other_column and pd.api.types.is_numeric_dtype(data[column]) and pd.api.types.is_numeric_dtype(data[other_column]):
          plt.plot_date(data[column], data[other_column], 'rs')
          plt.xlabel(column)
          plt.ylabel(other_column)
          plt.title(f'{column} vs {other_column}')
          plt.show()
          break # only plot one scatter for each column


# Pairplot for all columns
sns.pairplot(data)
plt.show()

# Pairplot for a subset of columns
sns.pairplot(data.iloc[:, 1:3])
plt.show()

warnings.filterwarnings('ignore')

# Distribution plot for all columns
for column in data.columns:
  if pd.api.types.is_numeric_dtype(data[column]):
    sns.distplot(data[column])
    plt.title(f'Distribution of {column}')
    plt.show()
```

```python
# prompt: can you create a bar plots on universities  data

# Assuming 'data' DataFrame is already loaded as in the previous code

# Create bar plots for specific columns
columns_to_plot = ['SAT', 'Top10', 'Accept']  # Example columns

for column in columns_to_plot:
  plt.figure(figsize=(8, 6))  # Adjust figure size as needed
  sns.barplot(x=data.index, y=data[column])
  plt.xlabel('University Index')  # Or a more descriptive x-axis label
  plt.ylabel(column)
  plt.title(f'Bar plot of {column} for Universities')
  plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for better readability
  plt.tight_layout() # Adjust layout to prevent labels from overlapping
  plt.show()
```
