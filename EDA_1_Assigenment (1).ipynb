```python
#  EXPLORATORY DATA ANALYSIS ON A DATASET
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Cardiotocographic.csv')
```

```python
# Task 1: Data Cleaning and Preparation
```

```python
# Check for missing values
```

```python
print("Missing values before imputation:\n", df.isnull().sum())
```

```python
# Impute missing values with the median
```

```python
df_filled = df.fillna(df.median())
```

```python
 # Check for missing values after imputation
```

```python
print("Missing values after imputation:\n", df_filled.isnull().sum())
```

```python
# Task 2: Statistical Summary
```

```python
# Generate statistical summary
```

```python
stat_summary = df_filled.describe().T
```

```python
# Calculate interquartile range (IQR) for each variable
```

```python
stat_summary['IQR'] = stat_summary['75%'] - stat_summary['25%']
```

```python
# Display key statistics
```

```python
print("Statistical Summary:\n", stat_summary[['mean', '50%', 'std', 'IQR']])
```

```python
# Task 3: Data Visualization
```

```python
# Set up the visualizations
plt.figure(figsize=(16, 12))
```

```python
# Histogram for 'LB' (Baseline Fetal Heart Rate)
```

```python
plt.subplot(1, 2, 1)
sns.histplot(df_filled['LB'], kde=True, color='skyblue')
plt.title('Baseline Fetal Heart Rate (LB) Distribution')
```

```python
# Boxplot for 'AC' (Accelerations)
```

```python
plt.subplot(1, 2, 1)
sns.boxplot(x=df_filled['AC'], color='lightgreen')
plt.title('Accelerations (AC) Distribution')
```

```python
# Histogram for 'FM' (Fetal Movements)
```

```python
plt.subplot(1, 2, 2)
sns.histplot(df_filled['FM'], kde=True, color='salmon')
plt.title('Fetal Movements (FM) Distribution')
```

```python
# Boxplot for 'UC' (Uterine Contractions)
plt.subplot(1, 2, 1)
sns.boxplot(x=df_filled['UC'], color='lightcoral')
plt.title('Uterine Contractions (UC) Distribution')
```

```python
# Histogram for 'ASTV' (Abnormal Short Term Variability)
plt.subplot(1, 2, 2)
sns.histplot(df_filled['ASTV'], kde=True, color='lightblue')
plt.title('Abnormal Short Term Variability (ASTV) Distribution')
```

```python
# Boxplot for 'MLTV' (Mean Long Term Variability)
plt.subplot(1, 2, 1)
sns.boxplot(x=df_filled['MLTV'], color='orange')
plt.title('Mean Long Term Variability (MLTV) Distribution')
```

```python
plt.tight_layout()
plt.show()
```

```python
# Task 4: Pattern Recognition and Insights
```

```python
# Correlation matrix and heatmap
plt.figure(figsize=(12, 8))
correlation_matrix = df_filled.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Pairplot for detailed relationships between variables
```

```python
sns.pairplot(df_filled)
plt.show()
```
